- sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=False); - scatterplot - overlapping plots

    - simple bar chart with single variable
    - still starts with zero
- sns.pointplot(auto['body_style'], auto['horsepower'], hue=auto['number_of_doors']);

## Plots and graphs

```python
import pandas as pd
import seaborn as sns # why sns? The West Wing series has a character name seaborn sam
import matplotlib.pyplot as plt
sns.set(color_codes=True) # adds a nice bg to the graphs
%matplotlib inline 
# tells python to render graphs instead of printing the object
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

## Plotting univariate distributions
The most convenient way to take a quick look at the univariate in seaborn is the distplot() function. Default histogram and fit kernal density estimate (KDE)

```python
sns.distplot(auto['highway_mpg']);
```

```python
sns.distplot(auto['city_mpg'], kde=False, rug=True); #turns off kde. rug is the tic mark along x axis
```

## Plotting bivariate distributions
Visualize relationship between two variables. joinplot() does a scatterplot of two variables along with histogram of each in the corresponding axis

```python
sns.jointplot(auto['engine_size'], auto['horsepower']); # semi colon is required hide object description 
```

### Hex bin plots
breaks 2d areas into hexagon and number  of points in each hex determines a color.
good for density

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="hex");
```

### Kernal Density Estimation
True density viz

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind='kde');
```

## Visualization pairwise relationships in dataset
To plot multiple pairwise scatterplots in a dataset, use pairplot() function. 
This creates a matrix of axes and shows the relationship for each pair of columns in data frame. also draws histogram of eachb variable on the diagonal axes

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']]);
```

## Plotting with categorical data (non-numbers) (vs continuous variables (numbers))

In a strip plot, the scatterplot points will overlap. This makes it difficult to see full distribution of data. jitter solves it - it adjusts the positions along categorical axis


```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True);
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=False);
```

Previous one had overlapping points. Use swarm plot to avoid overlapping

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower']);
```

## Box plots 
- similar to histogram
- line in middle of the box is the median; 
- q1(lower hor. line - 25% data below box), median, q3(upper hor. line - 75% data above box)
- whiskers extend to points that lie within 1.5 IQRs of the lower and upper quartile
- observations that fall outside this range are displayed independently (outlayers)

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower']); 
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type']);
```

## Bar plots
- uses bootstrapping to compute a confidence interval around the estimate and plots that using error bars.
- always start from zero

```python
sns.barplot(auto['body_style'], auto['horsepower'], hue=auto['fuel_type']);
```

```python
sns.countplot(auto['body_style']);
```

- Avoiding zero

```python
sns.pointplot(auto['body_style'], auto['horsepower'], hue=auto['number_of_doors']);
```

## Multi panel categorical plots


```python
sns.catplot(x="fuel_type", y="horsepower", col="drive_wheels", data=auto, kind="box");
```

```python
sns.catplot(x="fuel_type", y="horsepower", col="drive_wheels", data=auto, kind="violin"); # shows KDE
```
