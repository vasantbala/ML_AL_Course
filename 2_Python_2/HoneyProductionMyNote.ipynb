```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pp
```

```python
data = pd.read_csv('honeyproduction.csv')
```

```python
pp.figure(figsize=(20,10))
sns.countplot(data['state']);
```

```python
pp.figure(figsize=(20,10))
a = sns.countplot(data['state'])
a.set_xticklabels(a.get_xticklabels(), rotation=45);
```

```python
check=data['state'].value_counts().reset_index()
check.head()
```

```python
pp.figure(figsize=(20,10))
sns.barplot(check['index'], check['state']);
```
