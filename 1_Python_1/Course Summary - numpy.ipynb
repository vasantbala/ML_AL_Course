## NumPy
1. import numpy as np
2. Used for math operations that are more advanced than add, sub, mul, div
3. Has array and matrix data types. Unlike lists, values of arrays and matrix should be of single datatype

### Popular Functions in NumPy
1. np.cos(x)
2. np.pi
3. np.sqrt(x)
4. np.log(x)
5. np.exp(x)

### Arrays & Matrix

#### Creation
1. From List -> np.array([]), np.array([[],[]])
2. Transpose -> myMatrix.T
3. np.arange(f,l[,d]) -> returns array with a range defined by first(f) and last(l)(exclusive) with difference (d) b/w values
4. np.linspace(f,l,e) -> returns array with a range defined by first(f) and last(l)(inclusive) with equally space (e)
5. np.reshape(r,c) -> returns matrix from vector
6. np.zeros([r,c]) -> returns matrix with all zeros
7. np.ones((r,c)) -> returns matrix will all ones
8. np.eye(r) -> returns identity matrix. square matrix with value one on the diagonal and zero on others
9. np.random.rnand(r,c) -> returns matrix with uniform randon variable
10. np.random.randn(r,c) -> returns matric with std normal random variable
11. np.unique(array)
12. submatrix = matrix[rfi:rli, cfi:cli].copy()

#### Arithmetics
1. +, -, *, /, sqrt on arrays with same size. arithmetic applies to each values
array3 = array1 + array2 
array2 = 1/array1
2. np.matmul(mat, array)
3. np.linalg.solve(mat1, mat2)
4. np.linalg.inv(mat)
5. np.mean(matrix)
6. np.std(matrix)
7. np.min(matrix)
8. np.max(matrix)

#### Accessing

##### Array
1. By index -> array[i]
2. By range -> array[fi:li]
3. By non-consequtive indices -> array[np.array(i1, i4, i8. ...)]

##### Matrix
1. matrix[ri][ci] or matrix[ri,ci]
2. matrix[rf1:rli, cfi:rli]

#### Change Values
1. matrix[ri:ci] = newvalue
2. matrix[ri:ci] = [rvalue:colvalue]
3. matrix[rfi:rli, cfi:cli] = newvalue
4. matrix[:]=newvalue

#### with logical operators
1. array[array>0]
2. matrix[matrix>0]

1. np.save('filename', matrix)

3. np.savetxt('filename.txt', matrix, delimiter=',')
